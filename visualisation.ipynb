```python
y0 = 5
R = 3

t = np.linspace(0, 2 * 3.14, 100)

x = R * np.cos(t)
y = R * np.sin(t) - y0

plt.plot(x, y)

a1 = 10
a2 = 15

q2 = -np.arccos((x**2 + y**2 - a1**2 - a2**2)/(2*a1*a2))
q1 = np.arctan(y/x) + np.arctan(a2*np.sin(q2)/(a1+a2*np.cos(q2)))

def drow_angls(ang1, ang2):
    X_points = [0]
    Y_points = [0]

    y1 = -a1 * np.cos(ang1)
    x1 = a1 * np.sin(ang1)
    X_points.append(x1)
    Y_points.append(y1)


    y2 = -a2 * np.cos(ang2)
    x2 = a2 * np.sin(ang2)
    X_points.append(x2)
    Y_points.append(y2)

    plt.plot(X_points, Y_points)

    return X_points, Y_points

for ang1, ang2 in zip(q1, q2):
    X_points, Y_points = drow_angls(ang1, ang2)
```
